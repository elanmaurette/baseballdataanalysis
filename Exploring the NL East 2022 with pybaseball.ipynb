```python
import pybaseball as pyball
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
#A brief analysis and visualization of one of the most exciting divisions in 2022
# First we will get all NL East Team Records for 2022 Season

year = 2022
braves = pyball.schedule_and_record(year, 'ATL')
mets = pyball.schedule_and_record(year, 'NYM')
phillies = pyball.schedule_and_record(year, 'PHI')
marlins = pyball.schedule_and_record(year, 'MIA')
nationals = pyball.schedule_and_record(year, 'WSN')
```

```python
# Gets wins-to-date, losses-to-date, and win_percent-to-date for each team
mets['Wins'] = np.where(mets['W/L']=='W', 1, (np.where(mets['W/L']=='W-wo', 1, 0))).cumsum()
mets['Losses'] = np.where(mets['W/L']=='L', 1, (np.where(mets['W/L']=='L-wo', 1, 0))).cumsum()
mets['Win_Percentage'] = mets['Wins'] / (mets['Wins'] + mets['Losses'])

braves['Wins'] = np.where(braves['W/L']=='W', 1, (np.where(braves['W/L']=='W-wo', 1, 0))).cumsum()
braves['Losses'] = np.where(braves['W/L']=='L', 1, (np.where(braves['W/L']=='L-wo', 1, 0))).cumsum()
braves['Win_Percentage'] = braves['Wins'] / (braves['Wins'] + braves['Losses'])

phillies['Wins'] = np.where(phillies['W/L']=='W', 1, (np.where(phillies['W/L']=='W-wo', 1, 0))).cumsum()
phillies['Losses'] = np.where(phillies['W/L']=='L', 1, (np.where(phillies['W/L']=='L-wo', 1, 0))).cumsum()
phillies['Win_Percentage'] = phillies['Wins'] / (phillies['Wins'] + phillies['Losses'])

marlins['Wins'] = np.where(marlins['W/L']=='W', 1, (np.where(marlins['W/L']=='W-wo', 1, 0))).cumsum()
marlins['Losses'] = np.where(marlins['W/L']=='L', 1, (np.where(marlins['W/L']=='L-wo', 1, 0))).cumsum()
marlins['Win_Percentage'] = marlins['Wins'] / (marlins['Wins'] + marlins['Losses'])

nationals['Wins'] = np.where(nationals['W/L']=='W', 1, (np.where(nationals['W/L']=='W-wo', 1, 0))).cumsum()
nationals['Losses'] = np.where(nationals['W/L']=='L', 1, (np.where(nationals['W/L']=='L-wo', 1, 0))).cumsum()
nationals['Win_Percentage'] = nationals['Wins'] / (nationals['Wins'] + nationals['Losses'])
```

```python
# Graph Wins Comparison 
#Here we can see that Braves and Mets were in a close race all year long. 
#At about game 50 the Mets had a big lead but definitely not insurmountable. 
#What is really amazing is from that point on you see the Mets having a pretty steady performance, but the Braves got on a  
#Incredible roll after game 50 to keep it close all the way until the end. 

plt.rcParams["figure.figsize"] = (8,6)

plt.plot(braves['Wins'], label='Braves', c='red')
plt.plot(mets['Wins'], label='Mets', c='blue')
plt.plot(phillies['Wins'], label='Phillies', c='#b30000')
plt.plot(marlins['Wins'], label='Marlins', c='navy')
plt.plot(nationals['Wins'], label='Nationals', c='gold')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Wins')

plt.legend(loc='lower right')

plt.title("NL East Wins Comparison ({})".format(year))
```

```python
# Graph Win Percentage Comparison 
#First we create a modified column to numerically represent 'Games Back'

plt.rcParams["figure.figsize"] = (8,6)

plt.plot(braves['Win_Percentage'], label='Braves', c='red')
plt.plot(mets['Win_Percentage'], label='Mets', c='blue')
plt.plot(phillies['Win_Percentage'], label='Phillies', c='#b30000')
plt.plot(marlins['Win_Percentage'], label='Marlins', c='navy')
plt.plot(nationals['Win_Percentage'], label='Nationals', c='gold')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Win Percentage')

plt.legend(loc='lower right')

plt.title("NL Central Win Percentage Comparison ({})".format(year))
```

```python
# Insert new column called 'int_GB' that represents GB as either negative (meaning behind) or 0.0 meaning in first or tied for first
def modify_gb(gb):
    if gb == 'Tied':
        gb = 0.0
    elif 'up' in gb:
        gb = 0.0 # float(gb.replace('up ', ''))
    else:
        gb = -float(gb)
    
    return gb
    
braves['int_GB'] = braves['GB'].dropna().apply(modify_gb)
mets['int_GB'] = mets['GB'].dropna().apply(modify_gb)
phillies['int_GB'] = phillies['GB'].dropna().apply(modify_gb)
marlins['int_GB'] = marlins['GB'].dropna().apply(modify_gb)
nationals['int_GB'] = nationals['GB'].dropna().apply(modify_gb)
```

```python
#Graph NL East Games Back Comparison 
#What the Graph Shows
#Graphing 'Games Back' gives an even clearer image of how the NL East was contended throughout the year.
#The Nationals fell to an early 10 Games Back, and never recovered. 
#The Mets built an early commanding lead over the competition, reaching the highest point at the 50 game mark.
#Phillies, Marlins and Braves were pretty closed together until that same 50 game point where they started to make the climb to the top.
#The division was led by the Mets the whole way until the end, when they met the Braves with 3 games back for a division defining series.
#The braves would go on to defeat the Mets and sweep the Mets, trying the division and holding the Tie-Breaker,
#To become the NL East Champions.

plt.rcParams["figure.figsize"] = (8,6)

plt.plot(braves['int_GB'], label='Braves', c='red')
plt.plot(mets['int_GB'], label='Mets', c='blue')
plt.plot(phillies['int_GB'], label='Phillies', c='#b30000')
plt.plot(marlins['int_GB'], label='Marlins', c='navy')
plt.plot(nationals['int_GB'], label='Nationals', c='gold')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.yticks(np.arange(0, -30, step=-2))
plt.xlabel('Game')
plt.ylabel('Games Back (in negative)')

plt.title("NL East Games Back Comparison ({})".format(year))
plt.legend(loc='lower left')
```

```python
# Get total runs-to-date and total runs_allowed-to-date for each team
braves['Total_Runs'] = braves['R'].cumsum()
braves['Total_RA'] = braves['RA'].cumsum()

mets['Total_Runs'] = mets['R'].cumsum()
mets['Total_RA'] = mets['RA'].cumsum()

phillies['Total_Runs'] = phillies['R'].cumsum()
phillies['Total_RA'] = phillies['RA'].cumsum()

marlins['Total_Runs'] = marlins['R'].cumsum()
marlins['Total_RA'] = marlins['RA'].cumsum()

nationals['Total_Runs'] = nationals['R'].cumsum()
nationals['Total_RA'] = nationals['RA'].cumsum()
```

```python
#Graph Runs and Runs Allowed Comparisons illustrates some of the causal factors in each teams divisional rank.
#The Runs' graph shows that the Braves, Mets and Phillies offenses were scoring runs at a pretty similar rate.

#The Phillies were scoring Runs at a rate similar to the Mets and Braves, most season long. 
#However, they were the worst in Runs Allowed by a large margin. By the end of the season, the Phillies were the 2nd worst in Runs Allowed
#While they scored at comparable rates to the rest of the division, their pitching didn't fare so well.

#The Mets were very consistent the whole year, you don't see big drops on offense or pitching, something that is very rare and hard to do. 
#They consistently outpitched the entire division right up to the end of the season, when they were overtake by the Braves.

#This division is the story of the Braves season. They were playing catch up since the beginnning, but is clear that at the 50 game mark something clicked. 
#They got Ronald Acuña Jr back and aided by a 14 game winning streak they put the foot on the gas pedal and started batting and pitching better than the competition. 
#At the 50 game mark the went from 4th to 1st on Runs and Runs allowed on the division, overtaking even the consistent pace of the Mets. Amazing! 

#People like to say that the Mets crumbled under pressure, but data tells another story. 
#Mets stayed consistent all year, they just lost to a once on a lifetime performance by the Braves. 
#You have only thing left to do, tip your cap!

plt.rcParams["figure.figsize"] = (16,8)

plt.subplot(1,2,1)
plt.plot(braves['Total_Runs'], label='Braves', c='red')
plt.plot(mets['Total_Runs'], label='Mets', c='blue')
plt.plot(phillies['Total_Runs'], label='Phillies', c='#b30000')
plt.plot(marlins['Total_Runs'], label='Marlins', c='navy')
plt.plot(nationals['Total_Runs'], label='Nationals', c='gold')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs')

plt.title("NL East Runs Comparison ({})".format(year))
plt.legend(loc='lower right')

plt.subplot(1,2,2)
plt.plot(braves['Total_RA'], label='Braves', c='red')
plt.plot(mets['Total_RA'], label='Mets', c='blue')
plt.plot(phillies['Total_RA'], label='Phillies', c='#b30000')
plt.plot(marlins['Total_RA'], label='Marlins', c='navy')
plt.plot(nationals['Total_RA'], label='Nationals', c='gold')

plt.xticks(np.arange(0, len(braves.index), step=10))
plt.xlabel('Game')
plt.ylabel('Runs Allowed')

plt.title("NL East Runs Allowed Comparison ({})".format(year))
plt.legend(loc='lower right')
```

```python
#Graphing Total Home Attendance
braves['Total_Home_Attendance'] = braves[braves['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
mets['Total_Home_Attendance'] = mets[mets['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
phillies['Total_Home_Attendance'] = phillies[phillies['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
marlins['Total_Home_Attendance'] = marlins[marlins['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
nationals['Total_Home_Attendance'] = nationals[nationals['Home_Away'] == 'Home']['Attendance'].dropna().cumsum()
```
